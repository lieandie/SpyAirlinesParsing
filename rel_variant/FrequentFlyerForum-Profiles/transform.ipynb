```python
class Place:
    def __init__(self, country, city, airport):
        self.country = country
        self.city = city
        self.airport = airport
        
    def __eq__(self, other):
        return self.country == other.country and self.city == other.city and self.airport == other.airport
    
    def __hash__(self):
        return hash(self.country) ^ hash(self.city) ^ hash(self.airport)
```

```python
places_set = set([])
places = []

k = 0
for account_obj in forum:    
    for flight in account_obj['Registered Flights']:
        place = Place(
            flight['Arrival']['Country'],
            flight['Arrival']['City'],
            flight['Arrival']['Airport']
        )
        places_set.add(place)
        
        place = Place(
            flight['Departure']['Country'],
            flight['Departure']['City'],
            flight['Departure']['Airport']
        )
        places_set.add(place)
        
        k+=2

print(k, len(places_set))

places.append([
    'id',
    'country',
    'city',
    'airport'
])

k = 1
places_list = list(places_set)
for i in places_list:
    place = []
    place.append(k)
    place.append(i.country)
    place.append(i.city)
    place.append(i.airport)
    places.append(place)
    k += 1
```

```python
class Flight:
    def __init__(self, flight, date, codeshare, arrival, departure):
        self.flight = flight
        self.date = date
        self.codeshare = codeshare
        self.arrival = arrival
        self.departure = departure
        
    def __eq__(self, other):
        return self.flight == other.flight and self.date == other.date and self.arrival == other.arrival and self.departure == other.departure
    
    def __hash__(self):
        return hash(self.flight) ^ hash(self.date) ^ hash(self.arrival) ^ hash(self.departure)
```
