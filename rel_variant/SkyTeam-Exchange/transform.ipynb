## Разделение файла SkyTeam-Exchange.yaml

Для того чтобы распарсить этот файл его надо порезать на файлы поменьше (он слишком большой, библиотека его не хавает как есть)

```python
import re
import yaml
import csv
import pandas as pd
from pprint import pprint
```

```python
fi = open("SkyTeam-Exchange.yaml")
fo = None
names = []
for line in fi:
    res = re.match(r"^'(\d{4}-\d{2}-\d{2})':$", line)
    if res:
        if fo:
            fo.close()
            
        name = res[1]
        fo = open('yaml_parts/' + name + '.yaml', mode='w')
        names.append(name)
    fo.write(line)
fo.close()
```

## Парсинг получившихся файлов

Разделяем все на 3 сущности:

- Полеты
- Бонусные карты
- Места в самолете

```python
flights = []
cards = []
sits = []
for name in names:
    yaml_part = open('./yaml_parts/' + name + '.yaml')
    obj = yaml.load(yaml_part)
    
#     flight_id = 0
    for key_flight, value_flight in obj[name].items():
#         flight_id += 1
        for key_card, value_card in value_flight['FF'].items():
            flight = {}
            flight['id'] = key_flight
            flight['date'] = name
            flight['from'] = value_flight['FROM']
            flight['status'] = value_flight['STATUS']
            flight['to'] = value_flight['TO']
            flight['card_id'] = key_card
            flights.append(flight)
            
    print(name)
    yaml_part.close()
    
```

```python
print(len(flights))
print(len(cards))
print(len(sits))
```

## Запись объектов в csv файлы

```python
flights_val = []
flights_val.append(list(flights[0].keys()))
for i in flights:
    flights_val.append(list(i.values()))
flights_df = pd.DataFrame(flights_val)
flights_df.to_csv('./csv/SkyTeam-Exchange.flight.csv', index=False, header=False)
```

```python
cards_val = []
cards_val.append(list(cards[0].keys()))
for i in cards:
    cards_val.append(list(i.values()))

cards_df = pd.DataFrame(cards_val)
cards_df.to_csv('./csv/SkyTeam-Exchange.card.csv', index=False, header=False)
```

```python
sits_val = []
sits_val.append(list(sits[0].keys()))
for i in sits:
    sits_val.append(list(i.values()))

sits_df = pd.DataFrame(sits_val)
sits_df.to_csv('./csv/SkyTeam-Exchange.sit.csv', index=False, header=False)
```
