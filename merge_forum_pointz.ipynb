```python
import pandas as pd
```

```python
forum_merged = pd.read_csv('assets/merged/FrequentFlyerForum-Profiles.merged.csv')
pointz_merged = pd.read_csv('assets/merged/PointzAggregator-AirlinesData.xml.merged.csv')
```

```python
forum_merged['date'] = forum_merged['date'].astype('datetime64[ns]')
```

```python
forum_merged.drop(['Unnamed: 0', 'account_id'], axis=1, inplace=True)
```

```python
forum_merged.head()
```

```python
pointz_merged.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
pointz_merged['date'] = pointz_merged['date'].astype('datetime64[ns]')
```

```python
pointz_merged.head()
```

```python
merged = pd.merge(forum_merged, pointz_merged, how='outer', left_on=['number','flight', 'date'], right_on=['number','flight_code', 'date'])
```

```python
merged.drop(['bonus_program'], axis=1, inplace=True)
```

```python
merged
```

```python
first_name_x_null = merged.first_name_x.isnull()
last_name_x_null = merged.last_name_x.isnull()
merged.loc[merged[first_name_x_null].index, 'first_name_x'] = merged.loc[merged[first_name_x_null].index, 'first_name_y']
merged.loc[merged[last_name_x_null].index, 'last_name_x'] = merged.loc[merged[last_name_x_null].index, 'last_name_y']
```

```python
merged.drop(['first_name_y', 'last_name_y'], axis=1, inplace=True)
```

```python
flight_null = merged.flight.isnull()
merged.loc[merged[flight_null].index, 'flight'] = merged.loc[merged[flight_null].index, 'flight_code']
```

```python
merged.drop(['flight_code'], axis=1, inplace=True)
```

```python
airport_arrival_null = merged.airport_arrival.isnull()
airport_departure_null = merged.airport_departure.isnull()
merged.loc[merged[airport_arrival_null].index, 'airport_arrival'] = merged.loc[merged[airport_arrival_null].index, 'arrival_code']
merged.loc[merged[airport_departure_null].index, 'airport_departure'] = merged.loc[merged[airport_departure_null].index, 'departure_code']
```

```python
merged.drop(['departure_code', 'arrival_code'], axis=1, inplace=True)
```

```python
merged.rename(columns={'first_name_x': 'first_name', 'last_name_x': 'last_name'}, inplace=True)
```

```python
merged.to_csv('assets/merged/forum_pointz.merged.csv')
```
